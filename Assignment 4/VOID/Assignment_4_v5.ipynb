```python
## Load labels
labels = pd.read_csv('/content/drive/My Drive/MSDS 458/Assignment 4 - Final/Data/train_labels.csv')
labels.head()
```

```python
## Load images
file_paths = [[fname, '/content/drive/My Drive/MSDS 458/Assignment 4 - Final/Data/train' + fname] for fname in labels['filename']]
file_paths
```

# EDA and Preprocessing

```python
## Confirm number of labels and images
print(len(labels))
print(len(file_paths))
```

```python
## Create dataframe of images
images = pd.DataFrame(file_paths, columns=['filename', 'filepaths'])
images.head()
```

```python
## Create dataframe of images and labels
train_data = pd.merge(images, labels, how = 'inner', on = 'filename')
train_data.head()     
```

## Resize and Encode Data

```python
#https://www.geeksforgeeks.org/python-opencv-cv2-imread-method/
#https://pythonexamples.org/python-opencv-cv2-resize-image/

data = []     

for i in range(len(train_data)):
    img_array = cv2.imread(train_data['filepaths'][i], cv2.IMREAD_GRAYSCALE)   
    new_img_array = cv2.resize(img_array, (100, 100))      

    if train_data['label'][i] == 'with_mask':
        data.append([new_img_array, 1])
    else:
        data.append([new_img_array, 0])
```
